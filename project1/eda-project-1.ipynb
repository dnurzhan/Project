````python
import pandas as pd
import matplotlib.pyplot as plt


# Загрузка данных из файла
data = pd.read_csv("foreign_exchange_rates.csv")
````

````python
# 1. Какова общая структура данных?
data.head()
````

````python
# 2. Какие значения принимают различные переменные и какова их распределение?
data.describe()
````

````python
# 3. Какие валюты представлены в данных?
# Вывод списка валют
currencies = data.columns[2:]
print("Список валют:", currencies)
````

````python
# 4. Какие временные интервалы (периоды) охватываются данными?
# Определение временных интервалов
start_date = data['Time Serie'].min()
end_date = data['Time Serie'].max()
print("Временные интервалы:", start_date, " - ", end_date)
````

````python
# 5. Есть ли пропущенные значения в данных?
# Проверка наличия пропущенных значений
missing_values = data.isnull().sum()
print("Пропущенные значения:", missing_values)
````

````python
# Вопрос 6: Как менялись курсы различных валют по отношению к доллару США в течение времени?

# Преобразование данных в числовой формат
data[currencies] = data[currencies].apply(pd.to_numeric, errors='coerce')
# Визуализация распределения значений каждой валюты по годам/месяцам
for currency in currencies:
    plt.figure()
    data.plot(x='Time Serie', y=currency, legend=False)
    plt.title(currency)
    plt.xlabel('Date')
    plt.ylabel('Exchange rate')
    plt.show()
````

````python
# Вопрос 7: Какие валюты имеют самые высокие и самые низкие курсы по отношению к доллару США?
max_rates = data[currencies].max()
min_rates = data[currencies].min()
print("Валюта с самым высоким курсом: ", max_rates.idxmax())
print("Валюта с самым низким курсом: ", min_rates.idxmin())
````

````python
# Вопрос 8: Какие валюты имеют наибольшую и наименьшую волатильность курса по отношению к доллару США?
volatility = data[currencies].std()
print("Валюта с наибольшей волатильностью: ", volatility.idxmax())
print("Валюта с наименьшей волатильностью: ", volatility.idxmin())
````

````python
# Вопрос 9: Какие валюты имеют наибольшую и наименьшую тенденцию к росту/падению курса по отношению к доллару США?
trend = data[currencies].diff().mean()
print("Валюта с наибольшей тенденцией к росту: ", trend.idxmax())
print("Валюта с наибольшей тенденцией к падению: ", trend.idxmin())
````
