````python
train_na = (train_data.isna().sum()/len(train_data)).sort_values(ascending=False)
test_na = (test_data.isna().sum()/len(test_data)).sort_values(ascending=False)

sum_na = pd.concat([train_na[train_na>0],test_na[test_na>0]],axis=1, sort=True)
sum_na.columns = ["Train","Test"]
sum_na.fillna(0).sort_values(by=["Train","Test"], ascending=False)
````

````python
replacements = {
    "LotFrontage" : train_data["LotFrontage"].mean(), 
    "Electrical": train_data["Electrical"].mode()[0],
    "MasVnrArea": 0,
}

for key, val in replacements.items():
    train_data[key].fillna(val, inplace=True)
    test_data[key].fillna(val, inplace=True)


train_data.fillna("None", inplace=True)
test_data.fillna("None", inplace=True)
train_data.drop(columns=["Id"],inplace=True)
````

````python
train_y = train_data["SalePrice"]
test_y = test_data["SalePrice"]

train_data.drop(columns=["SalePrice"], inplace=True)
test_data.drop(columns=["SalePrice"], inplace=True)

data = pd.concat([train_data, test_data])

labels = {"LandSlope", "ExterQual", 
                 "ExterCond", "BsmtQual",
                 "BsmtCond", "BsmtExposure",
                 "BsmtFinType1", "BsmtFinType2",
                 "HeatingQC", "CentralAir",
                 "KitchenQual", "Functional",
                 "FireplaceQu", "Fireplace",
                 "GarageFinish", "GarageQual",
                 "GarageCond", "PoolQC",
                 }

def label_encoding(df_data, labels):
    from sklearn.preprocessing import LabelEncoder
    data = df_data.select_dtypes(include=["object"]).copy()
    encoded_data = pd.DataFrame()
    for col in data.columns:
        if col in labels:
            encoder = LabelEncoder()
            encoded_data["enc_" + col] = encoder.fit_transform(data[col].astype(str))
    return encoded_data

obj_encoded = label_encoding(data, labels)
data[obj_encoded.columns] = obj_encoded

train_data = data.iloc[:len(train_data)]
test_data = data.iloc[len(test_data):]
````

````python
#train_y = train_data["SalePrice"]
#test_y = test_data["SalePrice"]

#train_data.drop(columns=["SalePrice"], inplace=True)
#test_data.drop(columns=["SalePrice"], inplace=True)

train_X = train_data.select_dtypes(exclude=["object"]).drop(columns="Id").copy()
test_X = test_data.select_dtypes(exclude=["object"]).drop(columns="Id").copy()
````

````python
from sklearn.preprocessing import StandardScaler
import sklearn.linear_model as linear_model
from sklearn.pipeline import make_pipeline
import sklearn.ensemble as ensemble 
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
````

````python
def train_and_eval(model, X, y):
    model = model.fit(X, y)
    print("Train RMSE: {:.2f} & R2: {:.3f}".format(mean_squared_error(np.exp(y), np.exp(model.predict(X)), squared=False), r2_score(y, model.predict(X))))
    return model
````

````python
X, y = train_X.values, np.log(train_y.values)


models = []
print("Multiple Linear Regression")
models.append(train_and_eval(linear_model.LinearRegression(), X, y))

print("Ridge Linear Regression")
models.append(train_and_eval(make_pipeline(StandardScaler(), linear_model.Ridge(alpha=1)), X, y))

print("RidgeCV Linear Regression")
models.append(train_and_eval(make_pipeline(StandardScaler(), linear_model.RidgeCV()), X, y))

print("SGDRegressor Linear Regression")
models.append(train_and_eval(make_pipeline(StandardScaler(), linear_model.SGDRegressor()), X, y))

print("RandomForest Regression")
models.append(train_and_eval(make_pipeline(StandardScaler(), ensemble.RandomForestRegressor(n_estimators=150)), X, y))

print("SVR Regression")
models.append(train_and_eval(make_pipeline(StandardScaler(), SVR(kernel="linear")), X, y))

print("AdaBoostRegressor Regression")
models.append(train_and_eval(make_pipeline(ensemble.AdaBoostRegressor(n_estimators=150, loss='square')), X, y))

print("BaggingRegressor Regression")
models.append(train_and_eval(make_pipeline(ensemble.BaggingRegressor(n_estimators=150, bootstrap=True)), X, y))

print("GradientBoostingRegressor Regression")
models.append(train_and_eval(make_pipeline(ensemble.GradientBoostingRegressor(learning_rate=0.05, n_estimators=150)), X, y))
````

````python
def inference(models, X_test, ids_test):
    y_preds = []
    
    for model in models:
        y_preds.append(model.predict(X_test))
    y_preds = np.array(y_preds)
    y_pred = y_preds.mean(axis=0)
    print(y_pred.shape)
    print(ids_test.shape)
    return pd.DataFrame({"Id":ids_test, "SalePrice": np.exp(y_pred)})

X, y = train_X.values, np.log(train_y.values)
X_test = test_data.select_dtypes(exclude=["object"]).drop(columns="Id").values
ids_test = test_data["Id"].values.astype(int)

pd_ = inference(models, X_test, ids_test)
pd_.to_csv("result.csv", index=False)
````
